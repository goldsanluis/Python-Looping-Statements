## Lab 2 Exercise 1
#### Input an integer, determine and display 'prime' if the integer is prime else display 'not prime'.

```python
num = int(input("Enter an integer: "))

# assume it is prime first
is_prime = True

if num <= 1:
    is_prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print("prime")
else:
    print("not prime")
```

## Lab 2 Exercise 2
#### compute for the cube of integers. 

```python
# Input the size of the array
size = int(input("Enter the size of the array : "))

# Input the elements of the array
elements = list(map(int, input("Enter the elements separated by space: ").split()))

# Display the cube of each element
for i in range(size):
    print(elements[i] ** 3)
```

## Lab 2 Exercise 3
#### Generate a hollow square based on the + integer input side length n.

```python
# Input the side length
n = int(input("Enter the side length of the square: "))

for i in range(n):
    if i == 0 or i == n - 1:
        print('x' * n)
    else:
        print('x' + ' ' * (n - 2) + 'x')
```

## Lab 2 Exercise 4
#### Generate an inverted right triangle of height n.

```python
# Input the height
n = int(input("Enter the height of the triangle: "))

for i in range(n, 0, -1):
    print('*' * i)
```
